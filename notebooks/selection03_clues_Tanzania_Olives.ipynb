Plotting allele trajectory (code based on clues traj file)

```python
input_filename = "../results/Tanzania_Olive_clues/snp_results/chrom1_snp187825928"

epochs = np.load('%s.epochs.npy'%(input_filename))
freqs = np.load('%s.freqs.npy'%(input_filename))
logpost = np.load('%s.post.npy'%(input_filename))

f,ax = plt.subplots(1,1)
f.set_size_inches(20,10)

plt.pcolormesh(epochs[:-1],freqs,np.exp(logpost)[:,:])
plt.axis((0,len(epochs[:-1]),0,1.0))
plt.ylabel('Allele frequency',fontsize=20)
plt.xlabel('Generations before present',fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

cbar = plt.colorbar()
cbar.ax.set_ylabel('Posterior prob.\n\n',rotation=270,fontsize=20,labelpad=40)
cbar.ax.tick_params(labelsize=18) 
```
