Setting up a script to convert recombination maps.

```python
recomb_map_path = "/home/eriks/baboondiversity/data/PG_panu3_recombination_map/mikumi_pyrho_genetic_map_chr{}.txt"
chrom = 20
```

```python
recomb_map = pd.read_csv(recomb_map_path.format(chrom), sep=" ")
recomb_map
```

```python
snpfile = pd.read_csv('../mosaic/baboon_pops/snpfile.20', sep=' ', names=["rsID", "chr", "distance",
                                                                           "position", "allele", "allele ?"])
```

```python
recomb_map.loc[recomb_map.position.isin(snpfile.position)]
```

```python
for chrom in range(1, 21):
    recomb_map = pd.read_csv(recomb_map_path.format(chrom), sep=" ")
    snpfile = pd.read_csv('../mosaic/baboon_pops/snpfile.{}'.format(chrom), sep=' ', names=["rsID", "chr", "distance",
                                                                           "position", "allele", "allele ?"])
    print(chrom)
    with open('../mosaic/baboon_pops/rates.{}'.format(chrom), 'w') as file:
        recomb_subset = recomb_map.loc[recomb_map.position.isin(snpfile.position)]
        file.write("{}\n".format(len(recomb_subset)))
        recomb_join = " ".join(["{}".format(x) for x in recomb_subset["Genetic_Map(cM)"]])
        pos_join = " ".join(["{}".format(x) for x in recomb_subset["position"]])
        file.write(pos_join+"\n")
        file.write(recomb_join+"\n")
```

```python
# chrX conversion
recomb_map = pd.read_csv(recomb_map_path.format("X"), sep=" ")
snpfile = pd.read_csv('../mosaic/baboon_pops/snpfile.30'.format(chrom), sep=' ', names=["rsID", "chr", "distance",
                                                                           "position", "allele", "allele ?"])
print(chrom)
with open('../mosaic/baboon_pops/rates.{}'.format("30"), 'w') as file:
    recomb_subset = recomb_map.loc[recomb_map.position.isin(snpfile.position)]
    file.write("{}\n".format(len(recomb_subset)))
    recomb_join = " ".join(["{}".format(x) for x in recomb_subset["Genetic_Map(cM)"]])
    pos_join = " ".join(["{}".format(x) for x in recomb_subset["position"]])
    file.write(pos_join+"\n")
    file.write(recomb_join+"\n")
```
